# Pandas Data Visualization Exercise



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3') # Importing dataframe names df3
%matplotlib inline 
```

```python
df3.info() # Exploring dataset
```

```python
df3.head()
```

```python
# 1) Building Scatterplot of column b Vs a
df3.plot.scatter(x="a", y="b", c="red", s=30, figsize=(12,3)) # x axis, y axis, color,dot size,figure size
```

```python
# 2) histogram of the 'a' column

df3["a"].plot.hist()
```

```python
# 3) setting style to histogram

plt.style.use('ggplot')
df3["a"].plot.hist(alpha=0.5,bins=25)
```

```python
# 4) Boxplot comparing the a and b columns.

df3[["a","b"]].plot.box()
```

```python
# 5) kde plot of the 'd' column 

df3["d"].plot.kde()
```

```python
# Increasing linewidth and making the linestyle dashed.

df3["d"].plot.kde(lw=3, ls="dashed")
```

```python
# 6) Area plot of all the columns for just the rows up to 30. 

df3.iloc[0:30].plot.area(alpha=0.3)
```

```python
# 7) To get legend outside of the plot

df3.iloc[0:30].plot.area(alpha=0.4)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
